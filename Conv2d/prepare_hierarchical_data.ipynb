```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Fresh_Start/all_items_per_shop_train_and_test.csv')
data
```

```python
items = pd.read_csv('../csv_folder/items.csv')
items
```

```python
df = pd.merge(data, items, left_on=['item_id'], right_on=['item_id'], how='inner')
df
```

```python
df = df.drop(columns=['item_name'])
```

```python
df = df.pivot_table(index = ['shop_id', 'item_category_id', 'item_id'],
                     values = ['item_cnt_day'],
                     columns = ['date_block_num'],
                     fill_value = 0,
                     aggfunc='sum')
df
```
